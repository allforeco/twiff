```python
# Define text
text = tweet.text

# Remove #'s
for item in tweet.hashtags[::-1]:
    text = text[:item['start']] + text[item['end']:]
print("#'s removed: {}".format(text))
    
# Remove invalid chars
import re
text = re.sub('[,:;]', '', text)
print('delimeter chars removed: {}'.format(text))

# Tokenize string
from torchtext.data import get_tokenizer
torch_tokenizer = get_tokenizer("basic_english")
tokens = torch_tokenizer(text)
print('tokenized text: {}'.format(tokens))

# Remove invalid tokens
from torchtext.data.utils import ngrams_iterator
ngrams = list(ngrams_iterator(tokens, 3))

# Date
import dateparser
for token in tokens:
    date = dateparser.parse(token, settings={'STRICT_PARSING':True})
    if date is not None: print('Date: {}'.format(date))
        
# Location
# from mordecai import Geoparser
# geo = Geoparser()
# geo.geoparse(text)
```
